```python
co2_aulesA = dadesco2_ffill.iloc[:,:27]
co2_aulesA.head()
```

```python
co2_aulesA2 = pd.melt(co2_aulesA, id_vars=['Date'], value_vars=['A-11','A-12','A-13','A-14','A-21','A-22','A-23','A-24','A-31','A-32','A-33','A-34','A-35','A-36','A-41','A-42','A-43','A-44','A-51','A-52','A-53','A-54','A-55','A-56','A-61','A-62'])
co2_aulesA2 = co2_aulesA2.rename(columns = {'variable':'Aula', 'value':'CO2'}) # unitats co2: ppm
co2_aulesA2.head()
```

```python
df = co2_aulesA2.copy()

# Calculate the mean temperature at each time instant
mean_temperatures = df.groupby('Date')['CO2'].transform('mean')

# Calculate the distance of each class from the mean temperature
df['Distance'] = abs(df['CO2'] - mean_temperatures)

# Define a function to subsample each group
def subsample_group(group):
    # Sort the group by distance from mean temperature
    sorted_group = group.sort_values(by='Distance')
    
    # Keep 2 points closest to the mean temperature
    closest_points = sorted_group.iloc[:1]
    
    # Keep distant points
    distant_points = sorted_group.iloc[-int(len(group) * 0.2):]  # Adjust the fraction as needed
    
    # Combine closest and distant points
    subsampled_group = pd.concat([closest_points, distant_points])
    
    return subsampled_group

# Apply the subsampling function to each group
subsampled_df = df.groupby('Date', group_keys=False).apply(subsample_group)
print(subsampled_df.size)
subsampled_df['outdoors_co2'] = 400
subsampled_df['max_co2'] = 1000
subsampled_df['Color'] = 'Comfort'
subsampled_df.loc[subsampled_df['CO2']>1000, 'Color'] = 'Unhealthy'
subsampled_df.loc[subsampled_df['CO2']<400, 'Color'] = 'Outdoor air'
subsampled_df.to_csv('co2subsampledETSAB2023_clean.csv', index = False)
```
