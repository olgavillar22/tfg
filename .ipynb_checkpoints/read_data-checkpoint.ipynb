```python
date_col = dadesco2['Date']
time_col = dadesco2['Time']
```

```python
def get_date(x):
    x = str(x)
    return x.split(' ')[0]

def get_time(x):
    x = str(x)
    return x.split(' ')[1]
```

```python
dadesco2['Date'] = date_col.apply(get_date)
dadesco2['Time'] = time_col.apply(get_time)
```

```python
dadesco2.isnull().sum(0)
# Alarming nans: col 3, col A-33(antic),
#col A-S.1[DS A], col LA-SMOD, col bib.PS1, 
#col DSC EC-MAQ, col DSC EC-SZ
```

```python
# We make sure there isn't any value different than 1000 in 3rd col
# We can drop this column bc it doesn't give any info
# Also drop 2 more cols with 7000 or 8000 nan
```

```python
dadesco2 = dadesco2.drop(columns = ['MaxCO2[UPC]-Concentració de diòxid de carboni (CO₂)[ppm]', 'Qaire DSA A-33 (antic)[DS A]-Concentració de diòxid de carboni (CO₂)[ppm]','Qaire DSB Bib. PS1[DS B]-Concentració de diòxid de carboni (CO₂)[ppm]'])
```

```python
dadesco2_ffill = dadesco2.fillna(method = 'ffill')
```

```python
dadesco2_bfill = dadesco2.fillna(method = 'bfill')
```

```python
dadesco2_ffill.isnull().values.any()
```

```python
def new_col_names(names):
    new_names = {'Date':'Date','Time':'Time'}
    names = names[2:]
    for name in names:
        new_names[name] = name[10:len(name)-45]
    return new_names
```

```python
new_names = new_col_names(dadesco2_ffill.columns.array)
new_names
```

```python
dadesco2_ffill = dadesco2_ffill.rename(columns = new_names)
```

```python
dadesco2_ffill.to_csv('co2ETSAB2023_clean.csv', index = False)
```
