```python
import streamlit as st
def groupby_selected_time(df, time_agg):
    if time_agg == 'Month':
        return df.groupby(df['Date'].dt.to_period('M')).mean().reset_index()
    elif time_agg == 'Day':
        return df.groupby(df['Date'].dt.to_period('d')).mean().reset_index()
    elif time_agg == 'Hour':
        return df.groupby(df['Date'].dt.to_period('h')).mean().reset_index()

temp = pd.read_csv('temperatureETSAB2023_clean.csv')
temp['Date'] = pd.to_datetime(temp['Date'], format='%Y-%m-%d %H:%M:%S')

# Add groupby button custom
groupby_options = ['Month', 'Day', 'Hour']
select_groupby_var2= st.selectbox("Choose the time aggregation2", groupby_options)
temp = groupby_selected_time(temp, select_groupby_var2)
temp['Date'] = temp['Date'].dt.to_timestamp()
temp.head()
```
