```python
# get df with temp of each class of edifici A
temp_aulesA = dades_ffill.iloc[:,:27]
temp_aulesA.head()
```

```python
temp_aulesA2 = pd.melt(temp_aulesA, id_vars=['Date'], value_vars=['A-11','A-12','A-13','A-14','A-21','A-22','A-23','A-24','A-31','A-32','A-33','A-34','A-35','A-36','A-41','A-42','A-43','A-44','A-51','A-52','A-53','A-54','A-55','A-56','A-61','A-62'])
temp_aulesA2 = temp_aulesA2.rename(columns = {'variable':'Aula', 'value':'Temperatura'})
temp_aulesA2.head()
```

```python
df = temp_aulesA2.copy()

# Calculate the mean temperature at each time instant
mean_temperatures = df.groupby('Date')['Temperatura'].transform('mean')

# Calculate the distance of each class from the mean temperature
df['Distance'] = abs(df['Temperatura'] - mean_temperatures)

# Define a function to subsample each group
def subsample_group(group):
    # Sort the group by distance from mean temperature
    sorted_group = group.sort_values(by='Distance')
    
    # Keep 2 points closest to the mean temperature
    closest_points = sorted_group.iloc[:2]
    
    # Keep distant points
    distant_points = sorted_group.iloc[-int(len(group) * 0.2):]  # Adjust the fraction as needed
    
    # Combine closest and distant points
    subsampled_group = pd.concat([closest_points, distant_points])
    
    return subsampled_group

# Apply the subsampling function to each group
subsampled_df = df.groupby('Date', group_keys=False).apply(subsample_group)
print(len(subsampled_df))
epochs_mapping = {
    1: 'heating',
    2: 'heating',
    3: 'heating',
    4: 'air-conditioning',
    5: 'air-conditioning',
    6: 'air-conditioning',
    7: 'air-conditioning',
    8: 'air-conditioning',
    9: 'air-conditioning',
    10: 'heating',
    11: 'heating',
    12: 'heating'
}
subsampled_df['epoch'] = subsampled_df['Date'].dt.month.map(epochs_mapping)
#subsampled_df['max_comfort'] = np.where(subsampled_df['epoch'] == 'air-conditioning', 25, 23)
#subsampled_df['min_comfort'] = np.where(subsampled_df['epoch'] == 'air-conditioning', 23, 21)
subsampled_df['max_comfort'] = 25
subsampled_df['min_comfort'] = 21
# Add column to get color (up/down comfort zone)
subsampled_df['Color'] = 'Comfort'
subsampled_df.loc[subsampled_df['Temperatura']>25, 'Color'] = 'Too warm'
subsampled_df.loc[subsampled_df['Temperatura']<21, 'Color'] = 'Too cold'
subsampled_df.to_csv('temperaturesubsampledETSAB2023_clean.csv', index = False)
```

```python
# DIFFERENCES BETWEEN FLOORS TEMPERATURES
floors = new_temp['Planta'].unique()
# Create an empty DataFrame to store temperature differences
reference_df = pd.DataFrame()
reference_df['Date'] = new_temp[new_temp['Planta']==1]['Date']
reference_df['Planta1'] = new_temp[new_temp['Planta']==1]['avg_temp'] #store actual temp in pl1 (ref)

# Set a reference floor
reference_floor = 1
# Calculate the difference in average temperature between each floor and the reference floor
for floor in floors:
    if floor != reference_floor:
        floor_name = f'Planta{floor}'
        reference_temp = new_temp[new_temp['Planta'] == reference_floor]['avg_temp'].values
        floor_temp = new_temp[new_temp['Planta'] == floor]['avg_temp'].values
        floor_diff = floor_temp - reference_temp
        reference_df[floor_name] = floor_diff

# Display the DataFrame with temperature differences
print(reference_df)
```
